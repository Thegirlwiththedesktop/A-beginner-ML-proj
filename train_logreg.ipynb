# Training data using LOGISTIC REGRESSION

```python
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from numpy import *

import warnings 
warnings.filterwarnings('ignore')
```

```python
import matplotlib.pyplot
%matplotlib inline 
import numpy as np
# Importing training data set
train_data=pd.read_csv('train.csv', delimiter=",", header = None)
X_data = train_data.iloc[:,0:17].astype(float)
Y_data = train_data.iloc[:,-1]
#X_data

cv_data = pd.read_csv('cv.csv', delimiter=",", header = None)
X_cv = cv_data.iloc[:,0:17].astype(float)
Y_cv = cv_data.iloc[:,-1]

#X_cv
```

Logistic Regression

We are going to begin by importing preprocessing from sklearn to scale the CV and Train data that we will use. Scaling is used to reduce variation in our data because it is a large data set and has a wide range


```python
from sklearn import preprocessing

#scaling data to reduce variation
X_data = preprocessing.scale(X_data)
X_cv = preprocessing.scale(X_cv)
```

```python
#Implementing logistic regression
##this matrix, given the features predict  where there is a purchase or not                                                                                

from sklearn.model_selection import cross_val_score,cross_val_predict
from sklearn import metrics
from sklearn.linear_model import LogisticRegression

logmodel = LogisticRegression()
logmodel.fit(X_data,Y_data)  #fit training data
prediction = logmodel.predict(X_cv)

print(prediction)
scores = logmodel.score(X_cv,Y_cv)
print(scores)
```

```python
# changing regularization values
for i in range(1,50):
    logmodel = LogisticRegression(C=i)
    logmodel.fit(X_data,Y_data)
    prediction = logmodel.predict(X_cv)
    print(prediction)
    scores = logmodel.score(X_cv,Y_cv)
    print(scores)
```

```python
#take the largest score
logmodel2 = LogisticRegression(C=50)
logmodel2.fit(X_data,Y_data)
prediction = logmodel2.predict(X_cv)
print(prediction)
score = logmodel2.score(X_cv,Y_cv)
print(score)


```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(Y_cv, prediction))
print(classification_report(Y_cv,prediction))
```

```python
from sklearn.linear_model import LogisticRegression
logit1=LogisticRegression()
logit1.fit(X_data,Y_data)
###Confusion matrix with sklearn
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score
confusion_matrix(logit1.predict(X_cv),Y_cv)
##Computing false and true positive rates
fpr, tpr,_=roc_curve(logit1.predict(X_cv),Y_cv,drop_intermediate=False)



plt.figure()
##Adding the ROC

##This curve shows the trade-off between sensitivity(TPR) and specificity(1-FPR). A curve closer to the diagonal of the ROC space 
#indicates a less accurate test of a classifier
plt.plot(fpr, tpr, color='red',
 lw=2, label='ROC curve')
##Random FPR and TPR
plt.plot([0, 1], [0, 1], color='blue', lw=2, linestyle='--')
##Title and label
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC curve')
plt.show()
```
