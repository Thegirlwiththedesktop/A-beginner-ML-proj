# Neural Network Training File 

### Importing all necessary libraries

```python
import pandas as pd
#import numpy as np
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from numpy import *
#from sklearn.externals import joblib
#import matplotlib.pyplot
%matplotlib inline 

# for scaling 
from sklearn import preprocessing

# for training 
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

#import warnings 
#warnings.filterwarnings('ignore')
```

### Importing a training and cross validating dataset


```python
# Importing training dataset
train_data = pd.read_csv('train.csv', delimiter=",", header = None)
X_train = train_data.iloc[:,0:17].astype(float)
Y_train = train_data.iloc[:,-1]
#print(X_train)
#print(Y_train)


# Importing CV dataset 
cv_data = pd.read_csv('cv.csv', delimiter=",", header = None)
X_cv = cv_data.iloc[:,0:17].astype(float)
Y_cv = cv_data.iloc[:,-1]
#print(X_cv)

```

### Scaling Data 

```python
# Scale both the train and cv data to reduce the varience between datapoints or factors 

std_scale = preprocessing.StandardScaler().fit(X_train)

X_train_scaled = std_scale.transform(X_train)
X_cv_scaled = std_scale.transform(X_cv)

#X_train_scaled
#X_cv_scaled
```

### Training Data 

#### Prediction 

```python
#Show convergence graphically 
#MLPClassifier:
#activation='relu', alpha=1e-05, batch_size='auto', beta_1=0.9, beta_2=0.999, early_stopping=False, epsilon=1e-08,
#hidden_layer_sizes=(5,), learning_rate='constant', learning_rate_init=0.001, max_iter=600, momentum=0.9, n_iter_no_change=10, 
#nesterovs_momentum=True, power_t=0.5,random_state=1, shuffle=True, solver='lbfgs', tol=0.0001,validation_fraction=0.1, 
#verbose=False, warm_start=False

nodes = nodes = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 51, 52, 53]
#nodes = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 100, 120, 280, 360, 400]
#nodes = [5,7,9,11,14,17,20,24,28,32,37,42,47,52,80,120,150,300]

for i in nodes:
    clf = MLPClassifier(solver='sgd', alpha=1e-5, hidden_layer_sizes=(i,), random_state=1, max_iter = 1000)
    clf.fit(X_train_scaled, Y_train) 
    plt.plot(clf.loss_curve_)
    plt.xlabel('Iteration #')
    plt.ylabel('Loss')
    plt.title('Hidden layer with '+str(i)+' nodes')
    plt.show()
```

#### Accuracy

```python
# Obtaining CV accuracy 
# Precision seems to be the highest when nodes are 53

nodes = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 51, 52, 53]
#nodes = [5,7,9,11,14,17,20,24,28,32,37,42,47,52,80,120,150,300]
print("# of nodes \tAccuracy")
for i in nodes:
    clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(i,), random_state=1, max_iter = 600)
    clf.fit(X_train_scaled, Y_train) 
    
    Y_pred = clf.predict(X_cv)
    Y_pred = list(map(float, Y_pred))
    Y_CV = list(map(float, Y_cv))
    
    print("")
    #print("")
    print(str(i)+" \t  \t"+str(accuracy_score(Y_pred, Y_cv)))
```

#### Print The highest accuracy 

```python
# Number of nodes with highest accuracy
print("Nodes \t  Accuracy")
print("")
print(str(53)+" \t  "+str(accuracy_score(Y_pred, Y_cv)))


```

#### Confusion Matrix 

```python
# A confusion matrix is a summary of prediction results on a classification problem.

# Plot normalized confusion matrix.
print(confusion_matrix(Y_cv, Y_pred))

 
```

#### Classification Report

```python
# Plot classification report for CV
print(classification_report(Y_cv, Y_pred))
```
